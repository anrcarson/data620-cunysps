## Data setup

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import requests
import io
import numpy as np
import collections
```

We imported the libraries. Pandas was crucial for our work as it allowed the data to be treated as a database, rather than as a dictionary or a comma separated list. Networkx's functionality, alongside matplotlib's pyplot and numpy, is primarily graphing. The libraries requests and io were used for bringing the data off the internet (Github) and into Python. Collections was used for going through the degrees and calculating degree distribution.

After this, it was a matter of identifying where the data was located, using the requests library to pull the content, and then having pandas read the content in utf-8 format into a CSV with the assistance of the io library making it all into strings beforehand.

```python
data_folder = "https://github.com/PLombardo811/baseballdatabank/blob/master/upstream/"
Team_url = data_folder + "Teams.csv?raw=true"
Team_content = requests.get(Team_url).content
Team_csv = pd.read_csv(io.StringIO(Team_content.decode('utf-8')))
```

## Data cleaning

A glimpse was taken at the first few lines of the data itself. There was minimal observable data cleaning that would be necessary for the task at hand.

```python
print(Team_csv.head())
```

The variable yearID had its values converted into strings.

```python
Team_csv["yearID"] = Team_csv["yearID"].apply(str)
```

```python
print(Team_csv.head())
```

The resulting data was in the format required for graphing and easier to work with.

## Making the graph

```python
Team_pairs = Team_csv[["yearID", "franchID"]].copy()
Team_pairs["edges"] = Team_pairs.apply(lambda x: (x["yearID"], x["franchID"]), axis=1)
```

We created a copy of the dataset using only the columns we needed, yearID and franchID. These columns denote the years a given team was in existence under a particular name, and the franchise ID for that name. Using these columns, we were able to create a new column in our new variable, `Team_pairs`, for the edges. These edges were recorded as tuples with the first item being the yearID and the second being the franchID.

```python
G = nx.DiGraph()
G.add_edges_from(Team_pairs["edges"].tolist())
G.add_nodes_from(Team_csv["yearID"].tolist()+Team_csv["franchID"].tolist())
```

To make the directed graph, the previously-created edges column was put it into list format, and then the yearID and franchID columns were put into list format and combined to make the nodes.

```python
pos = nx.spring_layout(G,k=1.55,iterations=80)
```

The positions of the nodes were based on the spring layout.

```python
pos_labels = {}
keys = pos.keys()
for key in keys:
    x, y = pos[key]
    pos_labels[key] = (x, y)
```

Labels were created for each of the nodes.

```python
for node in Team_csv["franchID"].tolist():
    G.node[node]["category"] = "franchise"
for node in Team_csv["yearID"].tolist():
    G.node[node]["category"] = "year"
color_map = {
    "franchise": "skyblue",
    "year": "pink"
}
```

Each node was also assigned a category based off of its column name, and a map for colors was created based on whether or not the given node was categorized as a franchise or a year.

Finally came the drawing of the plot, where each node was colored based off of the category described and the edges were semi-transparent so as to be able to follow their paths more clearly.

```python
plt.figure(figsize=(20,20))
nx.draw_networkx_labels(G, pos=pos_labels, font_color='Blue')
nx.draw_networkx_edges(G, pos, edge_color="Gray", alpha=0.3)
nx.draw_networkx_nodes(G, pos, node_color=[color_map[G.node[node]['category']] for node in G], node_size=300)
plt.axis('off')
plt.show()
```

## Finding the degree distribution
The degrees of each node is the number of connections it has to other nodes while the degree distribution represents the probability distribution of those degrees in the network.

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
degrees, count = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(degrees, count)

plt.title("Degree Distribution")
plt.ylabel("Count")
plt.xlabel("Degree")
```

## Finding the clustering coefficient
The clustering coefficient represents how much nodes cluster together.

```python
G_undirected = G.to_undirected() #must first create an undirected graph
nx.average_clustering(G_undirected)
```

## Finding the graph's diameter
The diameter of a network graph is the greatest path between two nodes.

```python
try:
    nx.diameter(G_undirected)
except Exception as e:
    print(e)
    print("There are a total of " + str(len(list(nx.isolates(G_undirected)))) + " isolated nodes")
```

```python
try:
    nx.diameter(G)
except Exception as e:
    print(e)
    print("There are a total of " + str(len(list(nx.isolates(G)))) + " isolated nodes")
```

So we can calculate the diameter of our graph only if it is converted to an undirected graph.  Our true, directed graph, does not have a diameter because our graph isn't strongly connected -- this means that some nodes are disconnected, making the diameter infinite. However, no nodes are isolated.
